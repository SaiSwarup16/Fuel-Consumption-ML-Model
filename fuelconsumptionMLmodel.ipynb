```python
#ML Model - traing model with indepedent variable as cylinder and dependent variable as Co2Emission
x = data[['CYLINDERS']].values  #indepedent variable
y = data['CO2EMISSIONS'].values #dependent variable
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.4 , random_state=50) 
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.3 , random_state=80) 
regr= LinearRegression()
regr.fit(x_train,y_train)
y_pred = regr.predict(x_test)
print(f"accuracy score of these model is:{r2_score(y_test,y_pred)}")
```
